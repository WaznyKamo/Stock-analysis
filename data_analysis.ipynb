```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
f_ind = pd.read_csv("Data/WiG20_fundamental_indicators.csv", parse_dates=True)
lpp = pd.read_csv("Data/Daily_data/lpp_d.csv", parse_dates=True)
```

```python
f_ind.head()
```

```python
for column in f_ind.columns:
    if column in ['Spółka', 'Kwartały'] or f_ind[column].dtype == np.float64:
        pass
    else:
#         print(column)
        f_ind[column] = f_ind[column].astype(str).str.replace(" ","")
        f_ind[column] = pd.to_numeric(f_ind[column], errors='coerce')
```

```python
f_ind[f_ind['Spółka']=='LPP']
```

```python
# Mapowanie kwartałów na miesiące
map_quarter_to_date = {
    'Q1': '05',
    'Q2': '08',
    'Q3': '11',
    'Q4': '02'
}

def quarter_to_date(quarter):
    year, quarter = quarter.split('/')
    month = map_quarter_to_date[quarter]
    # Jeśli Q4, to miesiąc przypada na luty następnego roku
    if quarter == 'Q4':
        year = str(int(year) + 1)
    return pd.to_datetime(f'{year}-{month}-01')

f_ind['Data'] = f_ind['Kwartały'].apply(quarter_to_date)

f_ind[f_ind['Spółka']=='LPP']
```

```python
lpp['Data'] = pd.to_datetime(lpp['Data'])
lpp.info()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(f_ind[f_ind['Spółka']=='LPP']['Data'], f_ind[f_ind['Spółka']=='LPP']['Wartość księgowa'], marker='o', linestyle='-')
plt.title('Zmiana wartości księgowej w czasie')
plt.xlabel('Data')
plt.ylabel('Wartość księgowa')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df = f_ind[f_ind['Spółka']=='LPP']
df.head()
```

```python
df['Data'] = pd.to_datetime(df['Data'])
lpp['Data'] = pd.to_datetime(lpp['Data'])
# Sortujemy dane
df = df.sort_values('Data')
lpp = lpp.sort_values('Data')

# Tworzymy wykres
fig, ax1 = plt.subplots(figsize=(12, 6))

# Oś Y po lewej - Wartość księgowa
ax1.plot(df['Data'], df['Wartość księgowa'], color='blue', label='Wartość księgowa')
ax1.set_xlabel('Data')
ax1.set_ylabel('Wartość księgowa', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Oś Y po prawej - Kurs LPP
ax2 = ax1.twinx()
ax2.plot(lpp['Data'], lpp['Zamkniecie'], color='red', label='Kurs LPP')
ax2.set_ylabel('Kurs LPP', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Formatowanie
plt.title('Porównanie: Wartość księgowa vs Kurs LPP')
fig.autofmt_xdate()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA

# Upewnij się, że kolumny 'Data' są datetime
df['Data'] = pd.to_datetime(df['Data'])
lpp['Data'] = pd.to_datetime(lpp['Data'])

# Sortuj dane
df = df.sort_values('Data')
lpp = lpp.sort_values('Data')

# Tworzymy wykres z 3 osiami Y
fig = plt.figure(figsize=(12, 6))
host = host_subplot(111, axes_class=AA.Axes)
plt.subplots_adjust(right=0.75)

par1 = host.twinx()  # druga oś Y
par2 = host.twinx()  # trzecia oś Y

# Przesunięcie trzeciej osi na prawo
offset = 60
new_fixed_axis = par2.get_grid_helper().new_fixed_axis
par2.axis['right'] = new_fixed_axis(loc='right', axes=par2, offset=(offset, 0))
par2.axis['right'].toggle(all=True)

# Etykiety osi
host.set_xlabel("Data")
host.set_ylabel("Wartość księgowa")
par1.set_ylabel("Kurs LPP")
par2.set_ylabel("Zysk operacyjny")

# Wykresy
p1, = host.plot(df['Data'], df['Wartość księgowa'], label="Wartość księgowa", color='blue')
p2, = par1.plot(lpp['Data'], lpp['Zamkniecie'], label="Kurs LPP", color='green')
p3, = par2.plot(df['Data'], df['Zysk operacyjny'], label="Zysk operacyjny", color='red')

# Legenda
host.legend(handles=[p1, p2, p3], loc='upper left')

# Formatowanie
host.grid(True)
fig.autofmt_xdate()
plt.title("Wartość księgowa, Kurs LPP i Zysk operacyjny w czasie")
plt.show()
```
